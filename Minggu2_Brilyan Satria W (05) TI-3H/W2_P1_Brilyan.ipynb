```python
import pandas as pd

data = 'Titanic-Dataset.csv' #loading data
df= pd.read_csv(data) # df(data frame) reading data
df.head() #displaying first 5 rows of data
```

```python
df.info() #displaying information about data (total data is 891, but some columns have missing values)
```

```python
df.isnull().sum()
```

```python
# Age - mean (we will change the missing values with the mean of the column Age)
#  inplace=True means that we will change the original data
#  df2 = df. If we change the value of df2 the df will also change since it share the same memory location, that's why we use inplace=True
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK" we will change the missing values with "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus we will change the missing values with the modus of the column Embarked
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

```python
df.head() #displaying first 5 rows of data after changing the missing values
```

```python
df.info() #the data is now clean, no missing values
```

```python
#saving the clean data to a new file
df.to_csv('Titanic-Dataset-Clean.csv', index=False) #index=False means that we don't want to save the index of the data
```
